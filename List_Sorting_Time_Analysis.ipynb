```python
import timeit # To calculate Time
import matplotlib.pyplot as plt # To plot graph

def sort_list(lst):
    return sorted(lst) # In-built function to sort list

list_sizes = [5000, 10000, 15000, 20000, 25000]
time_taken = []

for size in list_sizes:
    numbers_list = list(range(size, 0, -1))  # Creating a list of numbers in descending order
    time = timeit.timeit(lambda: sort_list(numbers_list), number=1)  # Measuring time taken for one sorting operation, here, lambda function is used tocreate a small function here (sort_list())
    # number = 1 indicates executing one operation at a time
    time_taken.append(time)

# Plotting the graph
plt.plot(list_sizes, time_taken, marker='o')
plt.xlabel('Number of Elements in List')
plt.ylabel('Time Taken (seconds)')
plt.title('Time taken to sort different list sizes')
plt.show()
```
